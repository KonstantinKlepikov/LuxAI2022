```python
def get_struct(ob):
    
    structure = set()
    
    if isinstance(ob, list):
        for i in ob:
            structure.add(get_struct(i))
            
    elif isinstance(ob, dict):
        for k, v in ob.items():
            if 'factory' in k:                
                structure.add(('factory', get_struct(v)))
            elif 'unit' in k:
                structure.add(('unit', get_struct(v)))
            else:
                structure.add((k, get_struct(v)))
                          
    elif isinstance(ob, int):
        structure.add(int)
        
    elif isinstance(ob, str):
        structure.add(ob)
            
    return tuple(structure)
```

```python
st = get_struct(replay['steps'])
```
